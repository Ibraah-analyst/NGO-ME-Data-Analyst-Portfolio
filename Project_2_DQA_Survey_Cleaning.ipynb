```python
LOWER_BOUND=10
UPPER_BOUND=40

df_clean = df_clean = df_raw[
    (df_raw['temperature_C'] >= LOWER_BOUND) &
    (df_raw['temperature_C'] <= UPPER_BOUND)
].copy()

rows_dropped = df_raw.shape[0] - df_clean.shape[0]
rows_dropped
```

```python
df_clean.shape
```

```python
df_waste_summary = df_clean.groupby('waste_category')['food_waste_kg'].sum().sort_values(ascending=False).reset_index()
highest_waste_category = df_waste_summary.iloc[0]['waste_category']
highest_total_waste = df_waste_summary.iloc[0]['food_waste_kg']

print("--- Total Waste by Category (kg) ---")
print(df_waste_summary)
print(f"\nHighest Waste Category: {highest_waste_category.upper()} ({highest_total_waste:.2f} kg)")
```

```python
import plotly.express as px

# Create a bar chart using the summary data
fig_waste = px.bar(
    df_waste_summary,
    x='waste_category',
    y='food_waste_kg',
    color='waste_category',  # Assign a different color to each category
    title='Total Food Waste Contribution by Category',
    labels={'food_waste_kg': 'Total Waste (kg)', 'waste_category': 'Category'},
    text_auto='.2s'  # Automatically add the number labels on top of the bars
)

# Update layout for a cleaner look
fig_waste.update_layout(showlegend=False)

fig_waste.show()
```

```python
# 1. Filter the clean DataFrame for only the highest waste category ('meat')
df_meat = df_clean[df_clean['waste_category'] == 'meat'].copy()

# 2. Calculate the total aggregated metrics for the 'meat' category
total_meat_waste = df_meat['food_waste_kg'].sum()
total_meat_staff = df_meat['kitchen_staff'].sum()

# 3. Calculate the Staff Efficiency Metric (Waste kg per Staff Member)
staff_efficiency = total_meat_waste / total_meat_staff

print(f"--- Staff Efficiency for MEAT Waste ---")
print(f"Total Meat Waste (kg): {total_meat_waste:.2f}")
print(f"Total Kitchen Staff Days: {total_meat_staff}")
print(f"STAFF EFFICIENCY: {staff_efficiency:.2f} kg of waste per staff member")
```

```python
# Group by staff experience and calculate the mean food waste
df_experience_summary = df_clean.groupby('staff_experience')['food_waste_kg'].mean().sort_values(ascending=False).reset_index()

print("--- Mean Food Waste (kg) by Staff Experience ---")
print(df_experience_summary)
```

```python
from scipy import stats

# 1. Isolate the food_waste_kg data for each experience level
waste_beginner = df_clean[df_clean['staff_experience'] == 'beginner']['food_waste_kg']
waste_intermediate = df_clean[df_clean['staff_experience'] == 'intermediate']['food_waste_kg']
waste_expert = df_clean[df_clean['staff_experience'] == 'expert']['food_waste_kg']

# 2. Perform the One-Way ANOVA Test
anova_result = stats.f_oneway(waste_beginner, waste_intermediate, waste_expert)

# 3. Print the results
print("--- ANOVA Test for Food Waste by Staff Experience ---")
print(f"F-statistic: {anova_result.statistic:.4f}")
print(f"P-value: {anova_result.pvalue:.4f}")
```

```python
# Calculate the Pearson correlation coefficient (r)
correlation = df_clean['food_waste_kg'].corr(df_clean['past_waste_kg'])

print("--- Correlation Coefficient (r) ---")
print(f"Correlation (r) between Current Waste and Past Waste: {correlation:.4f}")
```

```python
import plotly.express as px

# 1. Select only the relevant numerical features for the correlation matrix
numerical_features = [
    'meals_served',
    'kitchen_staff',
    'temperature_C',
    'humidity_percent',
    'past_waste_kg',
    'food_waste_kg'
]

# 2. Calculate the Pearson correlation matrix
correlation_matrix = df_clean[numerical_features].corr()

# 3. Create the correlation heatmap using Plotly Express
fig_heatmap = px.imshow(
    correlation_matrix,
    text_auto=True,  # Show the correlation value (r) on the map
    aspect="equal", # Keep cells square
    color_continuous_scale=px.colors.sequential.Inferno, # Choose a color scale
    title="Correlation Heatmap of Key Numerical Features (r)"
)

fig_heatmap.show()
```
